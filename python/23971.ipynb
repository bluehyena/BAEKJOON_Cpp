```python
while True:
    a,b,c = map(int, input().split())

    if a == 0 and b == 0 and c == 0:
        break
    if a == b == c:
        print("Equilateral")
    elif a + b <= c or b + c <= a or c + a <= b:
        print("Invalid")
    elif a == b or b == c or c == a:
        print("Isosceles")
    else:      
        print("Scalene")
```

```python
string = input().upper() # input upper 는 대문자로 바꿔줌
string_list = list(set(string))
tmp = []

for i in string_list:
    count = string.count(i)
    tmp.append(count)

if tmp.count(max(tmp))>= 2:
    print("?")
else:
    print(string_list[tmp.index(max(tmp))])
```

```python
import sys
input = sys.stdin.readline

M = int(input())

S = set()

for i in range(M):
    orders = input().split()

    order = orders[0]

    if len(orders) == 2:
        num = int(orders[1])

    if order == 'add':
        S.add(num)
    elif order == 'check':
        if num in S:
            print(1)
        else:
            print(0)
    elif order == 'remove':
        if num in S:
            S.remove(num)
    elif order == 'toggle':
        if num in S:
            S.remove(num)
        else:
            S.add(num)
    elif order == 'all':
        S = set(range(1,21))
    elif order == 'empty':
        S.clear()
```

```python
import sys

input = sys.stdin.readline

n, m = map(int, input().split())
map = [list(map(int, input.split())) for _ in range(n)]
chk = [[False] * m for _ in range(n)]

dy = [0,1,0,-1]
dx = [1,0,-1,0]

def bfs(y, x):
    rs = 1
    q = [(y,x)]
    while q:
        ey, ex = q.pop()
        for k in range(4):
            ny = ey + dy[k]
            nx = ex + dx[k]
            if 0 <= ny < n and 0 <= nx < m:
                if map[ny][nx] == 1 and chk[ny][nx] == False:
                    rs += 1
                    chk[ny][nx] = True
                    q.append((ny,nx))

    return rs

cnt = 0
maxv = 0

for j in range(n):
    for i in range(m):
        if map[j][i] == 1 and chk[j][i] == False:
            chk[j][i] = True
            cnt += 1
            maxv = max(maxv, bfs(j,i))

print(cnt)
print(maxv)
```
